```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.hubertiming.com/results/2024Lilac"
response = requests.get(url)
response.status_code
```

```python
page = response.text
soup = BeautifulSoup(page,"lxml")
soup.prettify()
```

```python
title = soup.title
print(title.text)
```

```python
table = soup.find("table",attrs={"id":"individualResults"})
table
```

```python
row = soup.find_all("tr")
row
```

```python
import re
result_list=[]
for row in row:
  cell = row.find_all('td')
  cell_to_string=str(cell)
  cell_sub = (re.sub(re.compile('<.*?>'),'',cell_to_string))
  result_list.append(cell_sub)

#convert results into df
results = pd.DataFrame(result_list)
results
```

```python
result_split = results[0].str.split(',',expand=True)
result_split.tail(5)
```

```python

result_split.tail(5)
```

```python
result_remove = result_split
result_remove.head(10)
```

```python
# Getting table name to set for columns name

table = soup.find("table",attrs={"id":"individualResults"})
header = table.find_all("th")
column_name = []
for row in header:
    column_name.append(row.text)
column_name

result_remove.columns=column_name
result_remove.head(10)
```

```python
result_remove["Place"] = result_remove["Place"].astype(str).str.replace("[", "")
result_remove["Name"] = result_remove["Name"].astype(str).str.replace("\r\n\r\n","")
result_remove["Gender Place"] = result_remove["Gender Place"].astype(str).str.replace("\r\n\r\n","")
result_remove["Age Group Place"] = result_remove["Age Group Place"].astype(str).str.replace("\r\n\r\n","")
result_remove["Gun Time"] = result_remove["Gun Time"].astype(str).str.replace("]","")
result_remove.head(15)
```

```python
import pandas as pd
import numpy as np
result_remove.replace("None",np.nan,inplace=True)
coulmns_to_clean = ["Name","Gender","Age","City","State","Chip Pace","Time","Gender Place","Age Group","Age Group Place","Time to Start","Gun Time"]
result_remove.dropna(subset=coulmns_to_clean,inplace = True)
result_remove.head(5)
```

```python
result_remove.reset_index(drop=True, inplace=True)
result_remove.index += 1
result_remove.head(5)
```
